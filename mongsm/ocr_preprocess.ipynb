```python
# Open the file in read mode
with open('md (1).txt', 'r') as file:
    # Read the contents of the file
    content = file.read()

# Print the contents of the file
print(content)
```

```python
import json
from openai import OpenAI, RateLimitError
import backoff
from tqdm import tqdm

client = OpenAI()


@backoff.on_exception(backoff.expo, RateLimitError, max_time=60)
def extract_text(text):
    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {
                "role": "system",
                "content": "Extract questions and choices. You must output in TSV format. Question should have question header and each choice should have its own header such as A, B, C, D, or E. Skip questions that are not multiple-choice. ",
            },
            {
                "role": "user",
                "content": f"{text}.",
            },
        ],
    )
    return response.choices[0].message.content.strip()
```

```python
qnc = extract_text(content)
```
